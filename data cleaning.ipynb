```python
import pandas as pd
import os
```

```python
# CSV file called combined_data.csv
for filename in os.listdir('./CSV Data'):
    print(filename)
    for file in os.listdir(f'./CSV Data/{filename}'):
        if file.endswith('.csv'):
            df = pd.read_csv(f'./CSV Data/{filename}/{file}')
            new_df = df[['Issue Type', 'Status', 'Priority', 'Summary', 'Original Time Estimated', 'Total Time Spent']].copy()
            new_df = new_df[:-1]
            new_df['Summary'] = new_df['Summary'].str.replace(r'WS-\d+', '', regex=True)
            new_df['Summary'] = new_df['Summary'].str.replace(r'PALM-\d+', '', regex=True)
            new_df['Summary'] = new_df['Summary'].str.replace(r'APO-\d+', '', regex=True)
            new_df['Summary'] = new_df['Summary'].str.replace(r'MED-\d+', '', regex=True)
            new_df['Summary'] = new_df['Summary'].str.replace(r'[^\w\s]', '', regex=True)
            new_df['Issue Type'] = new_df['Issue Type'].str.replace(r'[^\w\s]', '', regex=True)
            # add header if the file doesn't exist yet
            if not os.path.isfile('./Cleaned Data/combined_data.csv'):
                new_df.to_csv('./Cleaned Data/combined_data.csv', mode='a', header=True, index=False)
            else:
                new_df.to_csv('./Cleaned Data/combined_data.csv', mode='a', header=False, index=False)
```

```python
# load cleaned data into a dataframe
project_df = pd.read_csv('./Cleaned Data/combined_data.csv')
```

```python
# remove all the records with the total timespent is empty ie '-'
project_df = project_df[project_df['Total Time Spent'] != '-']
```

```python
# remove all the records with the original time estimated is empty ie '-'
project_df = project_df[project_df['Original Time Estimated'] != '-']
```

```python
project_df.describe()
```

```python
def to_minutes(time_str: str, hours_per_day: int = 8) -> int:
    total_minutes = 0

    # Split the input string by spaces
    parts = time_str.split()

    for part in parts:
        if part.endswith('d'):
            num = float(part[:-1])
            total_minutes += num * hours_per_day * 60
        elif part.endswith('h'):
            num = float(part[:-1])
            total_minutes += num * 60
        elif part.endswith('m'):
            num = float(part[:-1])
            total_minutes += num

    return int(total_minutes)
```

```python
project_df['Total Time Spent'] = project_df['Total Time Spent'].apply(to_minutes)
```

```python
project_df['Original Time Estimated'] = project_df['Original Time Estimated'].apply(to_minutes)
```
